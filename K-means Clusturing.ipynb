```python
# Create a figure with two subplots (one the above the other)
fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios":(.15, .85)})

# Creating a boxplot
sns.boxplot(data=univ, x="SAT", ax=ax_box, orient='h')
ax_box.set(xlabel=' ') #Removes x label for the boxplot

#creating a histogram in th esame x-axis
sns.histplot(data=univ, x="SAT", ax=ax_hist, bins=30,kde=True, stat="count")
ax_hist.set(ylabel='count')

# Adjust layout'
plt.tight_layout()
plt.show()
```

#### Observations:
- it is a left-skewed plot
- in between 1100 to 1200 the universities are low
- and in between 1300-1400 the universities are high and majority is in between 1300-1400
